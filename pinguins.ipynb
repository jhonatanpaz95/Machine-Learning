```python
import numpy as np
import pandas as pd
import seaborn as sns

```

Base de dados com informações sobre penguins. 
A idéia é preparar a base de dados para prever a espécie do penguin (variável resposta) baseado em suas características físicas e geográficas (variáveis preditivas).

```python
data = sns.load_dataset('penguins')
data.shape
```

```python
# Exluindo os valores nulos

data.isnull().sum() # verifica a quantidade de valores nulos
data.dropna(inplace=True) # Exclui as linhas com valores nulos
data.shape
```

Padronização das variáveis Numéricas

```python
#df_num = data.select_dtypes(include=['float64', 'int64']) #seleciona so as colunas com valores numericos
#df_num['bill_length_mm_std'] = (df_num['bill_length_mm'] - df_num['bill_length_mm'].mean()) / df_num['bill_length_mm'].std()
```

```python
df_num = data.select_dtypes(include=['float64', 'int64'])  # Seleciona apenas colunas numéricas

lista_colunas = ['bill_length_mm', 'bill_depth_mm', 'flipper_length_mm', 'body_mass_g'] # lista das colunas para tratamento

for column in lista_colunas:
    df_num[f'{column}_std'] = (df_num[column] - df_num[column].mean()) / df_num[column].std()

df_num.drop(columns=lista_colunas, inplace=True)

df_num
```

```python

cat_df = data[['island','sex']]

cat_df['island_Dream_nom'] = cat_df['island'].apply(lambda x: 1 if x == 'Dream' else 0)
cat_df['island_Torgersen_nom'] = cat_df['island'].apply(lambda x: 1 if x == 'Torgersen' else 0)
cat_df['island_Biscoe_nom'] = cat_df['island'].apply(lambda x: 1 if x == 'Biscoe' else 0)
cat_df['sex_Male_nom'] = cat_df['sex'].apply(lambda x: 1 if x == 'Male' else 0)
cat_df['sex_Female_nom'] = cat_df['sex'].apply(lambda x: 1 if x == 'Female' else 0)


cat_df.drop(columns=['island', 'sex'], inplace=True)
cat_df.head()
```

```python
pinguins_df = pd.concat([data['species'], df_num, cat_df], axis=1)
pinguins_df.head()
```
